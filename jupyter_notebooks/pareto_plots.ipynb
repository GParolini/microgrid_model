```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "pareto_f_uncons.txt"

lines = open(file, "r").readlines()

lst=[]
for line in lines:
    line = line.split("\t")
    lst.append([line[0], line[1], line[2]])

for el in lst:
    el[0]=int(el[0].strip("point"))
    el[2]=el[2][:-1]

from operator import itemgetter

final_lst = sorted(lst, key=itemgetter(0))
print(final_lst)
```

```python
df = pd.DataFrame(final_lst, columns = ['Pareto_Point','Obj', 'Value']) 
df
```

```python
df_costs = df[df["Obj"]=="costs"]
df_emissions = df[df["Obj"]=="emissions"]
```

```python
df_costs.set_index("Pareto_Point")
```

```python
df_emissions.set_index("Pareto_Point")
```

```python
x = [round(float(x)) for x in df_costs.Value.to_list()]
y = [round(float(y)) for y in df_emissions.Value.to_list()]

plt.figure(figsize=(10, 8), dpi=300)
plt.scatter(x, y)
plt.grid()
plt.xlabel("Costs")
plt.ylabel("Greenhouse-Gas Emissions")

plt.show()
```
